```python
from matplotlib import pyplot as plt
import numpy as np

plt.rcParams['figure.dpi'] = 250
```

```python
# 70% with concatenation: accuracy, recall, precision, F1

models = ['MLP', 'GCN', 'GAT', 'SAGE', 'GATv2', 'GIN']

MLP = [78.43, 45.00, 78.26, 57.14]
GCN = [90.20, 100.0, 80.77, 90.14]
GAT = [90.20, 95.24, 83.33, 88.89]
SAGE = [92.16, 90.48, 90.48, 91.90]
GATv2 = [90.20, 100.0, 80.77, 90.14]
GIN = [90.20, 80.95, 94.44, 87.17]
```

```python
data = [MLP, GCN, GAT, SAGE, GATv2, GIN]
data = np.array(data)

bar_width = 0.2
x = np.arange(len(models))  # Base positions for the bars

fig, ax = plt.subplots(figsize=(10, 6))
metrics = ["Accuracy", "Recall", "Precision", "F1"]

colors = ["#B8B8B8", "#F2C45F", "#298C8C", "#800074"]
for i in range(data.shape[1]):
    ax.bar(x + i * bar_width, data[:, i], bar_width, label=metrics[i], color=colors[i])


ax.set_xlabel("Models")
ax.set_ylabel("Scores")
ax.set_title("Model Performance using 70% Training Data with Concatenation of News Data")
ax.set_xticks(x + bar_width * (data.shape[1] - 1) / 2)
ax.set_xticklabels(models)
ax.legend(title="Metrics")

plt.tight_layout()
plt.show()
```

```python
# 20%, 20% with concatenation, 70%, 70% with concatenation

gnns = ['GCN', 'GAT', 'SAGE', 'GATv2', 'GIN']

GCN_acc = [75.14, 81.92, 90.20, 90.20]
GAT_acc = [74.58, 80.23, 88.24, 90.20]
SAGE_acc = [78.53, 81.36, 94.12, 92.16]
GATv2_acc = [64.97, 73.45, 70.59, 90.20]
GIN_acc = [75.14, 79.66, 86.24, 90.20]
```

```python
acc_data = [GCN_acc, GAT_acc, SAGE_acc, GATv2_acc, GIN_acc]
acc_data = np.array(acc_data)

bar_width = 0.2
x = np.arange(len(gnns))  # Base positions for the bars

fig, ax = plt.subplots(figsize=(10, 6))
metrics = ["20% Training Data", "20% Training Data with Concatenation", "70% Training Data", "70% Training Data with Concatenation"]

colors = ["#B8B8B8", "#F2C45F", "#298C8C", "#800074"]
for i in range(acc_data.shape[1]):
    ax.bar(x + i * bar_width, acc_data[:, i], bar_width, label=metrics[i], color=colors[i])

ax.set_xlabel("GNN Models")
ax.set_ylabel("Accuracies")
ax.set_title("Comparison of GNN Test Accuracies")
ax.set_xticks(x + bar_width * (acc_data.shape[1] - 1) / 2)
ax.set_xticklabels(gnns)
ax.legend()
plt.legend(title="Columns", bbox_to_anchor=(1.05, 1), loc='upper left')

yticks = ax.get_yticks()
yticks = np.append(yticks, 100)
ax.set_yticks(sorted(yticks))

plt.tight_layout()
plt.show()
```
