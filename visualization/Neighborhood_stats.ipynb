```python
gnns = ["GCN", "GAT", "GraphSAGE", "GATv2", "GIN"]

GCN =  [86.27, 92.16, 92.16, 90.20]
GAT = [90.20, 88.24, 90.20, 90.20]
SAGE = [88.24, 90.20, 72.55, 92.16]
GATv2 = [90.20, 88.24, 90.20, 90.20]
GIN = [90.20, 90.20, 90.20, 90.20]
```

```python
hop_data = [GCN, GAT, SAGE, GATv2, GIN]
hop_data = np.array(hop_data)

bar_width = 0.2
x = np.arange(len(gnns))  # Base positions for the bars

fig, ax = plt.subplots(figsize=(10, 6))
hops = ["1", "2", "3", "No Limit"]

colors = ["#B8B8B8", "#F2C45F", "#298C8C", "#800074"]
for i in range(hop_data.shape[1]):
    ax.bar(x + i * bar_width, hop_data[:, i], bar_width, label=hops[i], color=colors[i])

yticks = ax.get_yticks()
yticks = np.append(yticks, 100)
ax.set_yticks(sorted(yticks))

ax.set_xlabel("GNN Models")
ax.set_ylabel("Accuracies")
ax.set_title("GNN Test Set Performance on Varying Neighborhood Sizes")
ax.set_xticks(x + bar_width * (hop_data.shape[1] - 1) / 2)
ax.set_xticklabels(gnns)
plt.legend(title="Max Distance to News Node", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
